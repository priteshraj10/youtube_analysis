```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="Nagpur Metro")

metro_stations = [
    'Khapri',
    'New Airport',
    'Airport South',
    'Airport Center',
    'Jaiprakash Nagar',
    'Ujjwal Nagar',
    'Chhatrapati Square',
    'Ajni',
    'Rahate Colony',
    'Congress Nagar',
    'Shankar Nagar',
    'Sitabuldi Interchange',
    'Zero Mile',
    'Lokmanya Nagar',
    'Bansi Nagar',
    'Vasudev Nagar',
    'Rachana Ring Road',
    'Ajni Square',
    'Tukdoji Bhavan',
    'Institute of Engineers',
    'Subhash Nagar',
    'Vasudeo Nagar',
    'Jhansi Rani Square',
    'Ajni Depot',
    'Automotive Square',
    'Wasudeo Nagar Depot'
]

coordinates = []

for station in metro_stations:
    location = geolocator.geocode(station + ", Nagpur, India")
    if location:
        coordinates.append((location.latitude, location.longitude))
    else:
        coordinates.append(None)

        # create a dataframe from to store latitude and longitude of each station

metro_df = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
metro_df['Station'] = metro_stations
metro_df
```

```python
import requests
import json
import pandas as pd
import time

# create a list to store distance and travel time between all metro stations
distance = []
time = []

for i in range(len(metro_df)):
    for j in range(len(metro_df)):
        if i != j:
            url = 'https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins={},{}&destinations={},{}&key={}'.format(
                metro_df.iloc[i]['Latitude'], metro_df.iloc[i]['Longitude'], metro_df.iloc[j]['Latitude'],
                metro_df.iloc[j]['Longitude'], 'YOUR_API_KEY_HERE')
            response = requests.get(url)
            data = json.loads(response.text)
            if data['status'] == 'OK':
                distance.append(data['rows'][0]['elements'][0]['distance']['text'])
                time.append(data['rows'][0]['elements'][0]['duration']['text'])
            else:
                print(f"No results for {metro_df.iloc[i]['Station']} to {metro_df.iloc[j]['Station']}")
                distance.append(None)
                time.append(None)
            time.sleep(0.5) # add a delay of half a second to avoid exceeding API usage limits
        else:
            distance.append(0)
            time.append(0)

# create a dataframe to store distance and travel time between all metro stations
distance_df = pd.DataFrame(distance, columns=['Distance'])
time_df = pd.DataFrame(time, columns=['Time'])
distance_df['Time'] = time_df['Time']
distance_df['Source'] = metro_df['Station']
distance_df['Destination'] = metro_df['Station']
distance_df = distance_df[['Source', 'Destination', 'Distance', 'Time']]
distance_df = distance_df.reset_index(drop=True)
distance_df

```
