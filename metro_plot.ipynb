```python
# Create a list to store all Nagpur Metro stations
metro_stations = ['Khapri', 'New Airport', 'Airport South', 'Airport Center', 'Jaiprakash Nagar', 'Ujjwal Nagar',
                  'Chhatrapati Square', 'Ajni', 'Rahate Colony', 'Congress Nagar', 'Shankar Nagar', 'Sitabuldi Interchange', 'Zero Mile',
                  'Lokmanya Nagar', 'Bansi Nagar', 'Vasudev Nagar', 'Rachana Ring Road', 'Ajni Square', 'Tukdoji Bhavan', 'Institute of Engineers',
                  'Subhash Nagar', 'Vasudeo Nagar', 'Jhansi Rani Square', 'Ajni Depot', 'Automotive Square', 'Wasudeo Nagar Depot'
                  ]
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
geolocator = Nominatim(user_agent="Nagpur Metro")

metro_stations = [
    'Khapri',
    'New Airport',
    'Airport South',
    'Airport Center',
    'Jaiprakash Nagar',
    'Ujjwal Nagar',
    'Chhatrapati Square',
    'Ajni',
    'Rahate Colony',
    'Congress Nagar',
    'Shankar Nagar',
    'Sitabuldi Interchange',
    'Zero Mile',
    'Lokmanya Nagar',
    'Bansi Nagar',
    'Vasudev Nagar',
    'Rachana Ring Road',
    'Ajni Square',
    'Tukdoji Bhavan',
    'Institute of Engineers',
    'Subhash Nagar',
    'Vasudeo Nagar',
    'Jhansi Rani Square',
    'Ajni Depot',
    'Automotive Square',
    'Wasudeo Nagar Depot'
]

coordinates = []

for station in metro_stations:
    location = geolocator.geocode(station + ", Nagpur, India")
    if location:
        coordinates.append((location.latitude, location.longitude))
    else:
        coordinates.append(None)

        # create a dataframe from to store latitude and longitude of each station

metro_df = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
metro_df['Station'] = metro_stations
metro_df

```

```python
# remove nan values
metro_df = metro_df.dropna()
metro_df
```

```python
# get distance between two cordinates using geopy library and store in a new dataframe metro_distance_df
import geopy.distance
metro_distance_df = pd.DataFrame(columns=['Station1', 'Station2', 'Distance'])
for i in range(len(metro_df)):
    for j in range(i+1, len(metro_df)):
        coords_1 = (metro_df.iloc[i]['Latitude'], metro_df.iloc[i]['Longitude'])
        coords_2 = (metro_df.iloc[j]['Latitude'], metro_df.iloc[j]['Longitude'])
        metro_distance_df = metro_distance_df.append({'Station1': metro_df.iloc[i]['Station'],
                                                      'Station2': metro_df.iloc[j]['Station'],
                                                      'Distance': geopy.distance.distance(coords_1, coords_2).km}, ignore_index=True)

metro_distance_df

# save metro_distance_df to csv file
metro_distance_df.to_csv('metro_distance.csv', index=False)
```

```python
# show the map as per the distance between stations
import folium
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap
from folium.plugins import FastMarkerCluster

# create a map of Nagpur
map_nagpur = folium.Map(location=[21.1458, 79.0882], zoom_start=12)

# add markers to map 
for lat, lng, station in zip(metro_df['Latitude'], metro_df['Longitude'], metro_df['Station']):
    label = '{}'.format(station)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_nagpur)

# add lines between stations as per the distance
for i in range(len(metro_distance_df)):
    if metro_distance_df.iloc[i]['Distance'] <= 2:
        folium.PolyLine(locations=[[metro_df[metro_df['Station'] == metro_distance_df.iloc[i]['Station1']].iloc[0]['Latitude'],
                                    metro_df[metro_df['Station'] == metro_distance_df.iloc[i]['Station1']].iloc[0]['Longitude']],
                        [metro_df[metro_df['Station'] == metro_distance_df.iloc[i]['Station2']].iloc[0]['Latitude'],
                         metro_df[metro_df['Station'] == metro_distance_df.iloc[i]['Station2']].iloc[0]['Longitude']]],
                        weight=1, color='red').add_to(map_nagpur)

map_nagpur



```
